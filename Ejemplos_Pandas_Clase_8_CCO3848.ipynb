# Ejercicios Pandas

## Filtrar datos de "contratos_iniciados_agosto_2021.xlsx", que tengan un "MONTO ADJUDICADO (\$)" menor a $150.000.000 y no sean de la región "Metropolitana de Santiago" o "de Valparaíso".

```python
# and = &
# or = |
```

```python
df_excel = pd.read_excel("contratos_iniciados_agosto_2021.xlsx")
```

```python
df_excel_filtrado = df_excel[df_excel["MONTO ADJUDICADO ($)"] < 150000000]
df_excel_filtrado = df_excel_filtrado[
    (df_excel_filtrado["REGION"] != "Metropolitana de Santiago")
    & (df_excel_filtrado["REGION"] != "de Valparaíso")]
df_excel_filtrado
```

## Agrupar el resultado según "REGION"

```python
df_agrupado = df_excel_filtrado
df_agrupado = df_agrupado.groupby(by=['REGION']).sum()
df_agrupado["MONTO ADJUDICADO ($)"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Lineplot

```python
df_excel = pd.read_excel("contratos_iniciados_julio_2021.xlsx")
```

```python
sns.lineplot(x="SERVICIO", y="MONTO ADJUDICADO ($)", data=df_excel)
sns.set(rc={'figure.figsize':(10,5)})
plt.xticks(rotation=90)
plt.show()
```

```python
df_excel[df_excel["SERVICIO"] == "AEROPUERTOS                        "]
```

```python
df_excel[df_excel["SERVICIO"] == "A.P.R. - COVID                     "]
```

## Violinplot

```python
sns.violinplot(data=df_excel, x="REGION", y="MONTO ADJUDICADO ($)")
sns.set(rc={'figure.figsize':(20,10)})
plt.xticks(rotation=90)
```

## Boxplot

```python
sns.boxplot(data=df_excel, x="REGION", y="MONTO ADJUDICADO ($)")
sns.set(rc={'figure.figsize':(20,10)})
plt.xticks(rotation=90)
```
